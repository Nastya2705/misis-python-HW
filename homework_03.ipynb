```python
def count_word_frequency(text):
    text = text.lower() #приводим текст к нижнему регистру
    
    marks = '.,!?;:' 
    for mark in marks:
        text = text.replace(mark,' ')  #удаляем знаки припенания
        
    text = text.split() #разделяем текст по пробелам
    
    dictionary = {}  #создаем пустой словарь

    for word in text:
        dictionary[word] = dictionary.get(word, 0) + 1 #считаем слова
        
    return(dictionary)
    
```

```python
# Тестируем функцию
text = input()
count_word_frequency(text)
```

```python
def calculate_price(price, discount=0, tax_rate=0.2):
    price_after_discount = price * (1 - discount / 100) # Применяем скидку
    final_price = price_after_discount * (1 + tax_rate) # применяем налог
    return final_price
    
```

```python
print(calculate_price(50))
```
