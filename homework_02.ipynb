```python
number = int(input())
if number <= 0:
    print('Число меньше или равно 0. Введите другое число.') 
for num in range(number):
    if (num + 1) % 3 == 0 and (num + 1) % 5 == 0:
        print('FizzBuzz')
    elif (num + 1) % 3 == 0:
        print('Fizz')
    elif (num + 1) % 5 == 0:
        print('Buzz')
    else:
        print((num + 1))
```

```python
shopping_list = []
while True:
    print(f'Меню \n 1. Добавить \n 2. Показать \n 3. Выйти')
    action = int(input('Выбирите цифру необходимого действия:'))
    if action == 1:
        product = input()
        shopping_list.append(product)
    elif action == 2:
        for index, products in enumerate(shopping_list):
            print(f'{index + 1}. {products}')
        if len(shopping_list) == 0:
            print('Список пуст')
    elif action == 3:
        print('Программа завершена.')
        break
    else:
        print('Вы что-то напутали. Попробуйте еще раз.')
        
        
    
```
